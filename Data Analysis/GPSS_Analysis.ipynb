```python
code = '50 CK'
mc_code = 'RF-01'
calc_df = calc_df[calc_df['pt_count'] == code]
cm_df = cm_df[cm_df['cm_count']==code]
std_gpss = cm_df.iloc[0,1]
```

```python
rf_ms = list(calc_df['pt_mcode'].unique())
```

```python
calc_df['act_gpss'].describe()
```

```python
# Efficiency Graph
months = []
for i in range(1,13):
    month = datetime.date(1900,i, 1).strftime('%B')
    months.append(month)
df_t = pd.DataFrame(months, columns = ['months']) 
df_t
```

```python
calc_df.reset_index()
calc_df['pt_date'] =pd.to_datetime(calc_df.pt_date)
for ms in rf_ms:
    monthly_data = []
    calc_df_t = calc_df[calc_df['pt_mcode'] ==  ms]
    grouped = calc_df_t.groupby('pt_date', as_index=False)
    grouped = grouped['act_gpss'].agg(np.mean)
  #  print(grouped)
    for i in range(1,13):
        mon_grouped_flags = grouped['pt_date'].map(lambda x: x.month) == i
        mon_grouped = grouped[mon_grouped_flags]
        monthly_data.append(np.mean(mon_grouped['act_gpss']))
    df_t[ms]=monthly_data
df_t.index = months
df_t.drop(['months'], axis=1, inplace=True)
df_t
```

```python
# ANNOTATE FOR ALL MACHINES STATICALLY
max_gpss = []
max_d={}
for col in df_t.columns:
    max = 0
    mon= 0
    vals = df_t[col]
    for i in range(len(vals)):
        if vals[i]>max:
            max = vals[i]
            mon = i
    max_d[max] = mon
max_d
```
