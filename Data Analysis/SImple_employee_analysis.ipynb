```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from matplotlib.pyplot import figure
import datetime
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
prod_t_df = pd.read_csv('pt.csv',low_memory=False)
```

```python
em_df = pd.read_csv('em.csv',low_memory=False)
em_df = em_df[['em_ecode','em_ename']]
```

```python
name_dict = {}
for index,row in em_df.iterrows():
    name_dict[row['em_ecode']] = row['em_ename']
```

```python
name_dict
```

```python
# Selecting RingFrame Machines in Code 'C'
unit_code = 'C'
prod_df = prod_t_df[['pt_date','pt_ucode','pt_count','pt_mcode','pt_hanks','pt_dlvrs','pt_stdhnk','pt_nomhnk','pt_idlespnper','pt_tc','pt_ecode1']]
prod_df = prod_df[prod_df['pt_ucode']== unit_code]
#prod_df = prod_df [prod_df.pt_mcode.str.contains('^RF')]
prod_df =  prod_df[prod_df.pt_date.str.contains('2018$')]
prod_df['pt_date'] =pd.to_datetime(prod_df.pt_date)
prod_df = prod_df.sort_values(by='pt_date')
prod_df.head(5)
```

```python
#Calculating Standard Production
prod_df['std_prod'] = (0.4536*prod_df['pt_stdhnk']*prod_df['pt_dlvrs'])/(prod_df['pt_nomhnk'])
prod_df.head(5)

# Calculating Actual Production
prod_df['act_prod'] =  ((0.4536*prod_df['pt_hanks']*prod_df['pt_dlvrs'])/(prod_df['pt_nomhnk']))*((100-prod_df['pt_tc'])/100)
prod_df.head(5)

#Calculating Production Efficiency
prod_df['prod_eff'] = (prod_df['act_prod']/prod_df['std_prod'])*100
prod_df.head(5)
```

```python
sorted(prod_df['prod_eff'],reverse=True)
```

```python
prod_df = prod_df.drop(['pt_ucode','pt_hanks','pt_dlvrs','pt_stdhnk','pt_nomhnk','pt_idlespnper','pt_tc','std_prod','act_prod'],axis = 1)
prod_df.columns
```

```python
prod_df = prod_df[prod_df['pt_ecode1']!='zzz']
prod_df = prod_df.replace([np.inf, -np.inf], np.nan)
prod_df = prod_df.dropna()
```

```python
months = []
for i in range(1,13):
    month = datetime.date(1900,i, 1).strftime('%B')
    months.append(month)
df_t = pd.DataFrame(months, columns = ['months']) 
df_t
```

```python
empcodes = []
empcodes.extend(list(prod_df['pt_ecode1'].unique()))
empcodes
prod_df.dtypes
```

```python
import multiprocessing as mp
import tqdm
num_processes = mp.cpu_count()
num_processes
```

```python
from itertools import repeat
from multiprocessing import Process, Manager
from multiprocessing import Pool
```

```python
# SPLIT DATAFRAME METHOD
import workers
num_partitions = num_processes
if __name__ ==  '__main__':
    manager = Manager()
    d = manager.dict()
    df_split = np.array_split(prod_df, num_partitions)
    pool = Pool(num_processes)
    a_args = repeat(d,num_partitions)
    for _ in tqdm.tqdm(pool.map(workers.process_rows, zip(a_args,df_split)), total=num_partitions):
        pass
    pool.close()
    pool.join()
```
