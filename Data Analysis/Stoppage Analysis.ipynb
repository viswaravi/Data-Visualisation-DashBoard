```python
# Importing Required libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from matplotlib.pyplot import figure
import datetime
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
stop_t_df = pd.read_csv('st.csv',low_memory=False,error_bad_lines=False)
```

```python
stop_t_df['st_reason'].unique()
```

```python
rf_code = 'RF-01'
count_code = '40s CKC'
unit_code = 'C'
```

```python
#Filtering Machine , Year and Unit Code and Sorting on Date
stop_data = stop_t_df[stop_t_df.st_mcode.str.contains('^RF')]
stop_data =  stop_data[stop_data.st_date.str.contains('^2018')]
stop_data = stop_data[stop_data['st_mcode']== rf_code] 
stop_data = stop_data[stop_data['st_count']== count_code]
stop_data = stop_data[stop_data['st_ucode']== unit_code]
stop_data['st_date'] =pd.to_datetime(stop_data.st_date)
stop_data = stop_data.sort_values(by='st_date')
#st=stop_data[(stop_data['st_date'] > '2018-06-01')]
stop_data
```

```python
#Grouping by Date
#grouped = stop_data.groupby(['st_date','st_reason'], as_index=False)

grouped = stop_data.groupby(['st_reason','st_date',], as_index=False)
grouped = grouped['st_rcode'].agg(['count'])
indexes = list(grouped.index)
dates = [ind[1] for ind in indexes]
dates = list(set(dates))
```

```python
data = []
reasons = [ind[0] for ind in indexes]
reasons = set(reasons)
print(reasons)
```

```python
months = []
for i in range(1,13):
    month = datetime.date(1900,i, 1).strftime('%B')
    months.append(month)
months
```

```python
#Summing Minthly Grouping
data = []
for rs in reasons:
    d_t = grouped.loc[rs]
    mon_data = []
    for i in range(1,13):
        mon_grouped_flags = d_t.index.map(lambda x: x.month) == i
        mon_grouped = d_t[mon_grouped_flags]
        mon_data.append(sum(mon_grouped['count']))
    data.append(go.Bar(
        x= months,
        y= mon_data,
        name=rs
    ))
layout = go.Layout(
    barmode='stack'
)
fig = go.Figure(data=data, layout=layout)
plot(fig)
```
