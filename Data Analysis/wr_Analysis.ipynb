```python
#MonthWise

month = 1
mon_grouped_flags = wr_df_t['wr_sdate'].map(lambda x: x.month) == month


mon_grouped = wr_df_t[mon_grouped_flags]


grouped = mon_grouped.groupby(['wr_sdate'], as_index=False)

grouped = grouped[['wr_acount','wr_stren','wr_rh']].agg(np.mean)


grouped['csp'] = grouped['wr_acount'] * grouped['wr_stren']


sample_group = mon_grouped.groupby(['wr_sdate'], as_index=False)
#print(sample_group.groups)
#print(mon_grouped)
sample_list = []
stren_cv_list = []
acount_cv_list = []
for name, group in sample_group:
    #print(set(list(group['wr_studyno'])))
    #print(group[['wr_studyno','wr_nosmp']])
    stren_cv = (np.std(group['wr_stren'])/np.mean(grouped['wr_stren']))*100
    acount_cv = (np.std(group['wr_acount'])/np.mean(grouped['wr_acount']))*100
    study_group = group.groupby(['wr_studyno'], as_index=False)
    study_group = study_group['wr_nosmp']. agg(np.max)
    #print(study_group)
    #print(sum(study_group['wr_nosmp']))
    sample_list.append(sum(study_group['wr_nosmp']))
    stren_cv_list.append(stren_cv)
    acount_cv_list.append(acount_cv)
```

```python
stren_cv_list = list(map(str, stren_cv_list))
stren_cv_list = ['strength_CV: '+ s for s in stren_cv_list]
stren_cv_list
```

```python
#len(acount_cv_list)
acount_cv_list = list(map(str, acount_cv_list))
acount_cv_list = ['count_CV: '+s for s in acount_cv_list]
acount_cv_list
```

```python
#YearWise Report
y_sample = []
count_list = []
stren_list = []
csp_list = []
rh_list = []
for m in range(1,13):
    print('MONTH',m)
    mon_grouped_flags = wr_df_t['wr_sdate'].map(lambda x: x.month) == m
    mon_grouped = wr_df_t[mon_grouped_flags]
    grouped = mon_grouped.groupby(['wr_sdate'], as_index=False)
    grouped = grouped[['wr_acount','wr_stren','wr_rh']].agg(np.mean)
    grouped['csp'] = grouped['wr_acount'] * grouped['wr_stren']
    sample_group = mon_grouped.groupby(['wr_sdate'], as_index=False)
    #print(sample_group.groups)
    print(grouped)
    count_list.append(np.nanmean(grouped['wr_acount']))
    stren_list.append(np.nanmean(grouped['wr_stren']))
    csp_list.append(np.nanmean(grouped['csp']))
    rh_list.append(np.nanmean(grouped['wr_rh']))
    sample_list = []
    for name, group in sample_group:
        # print(set(list(group['wr_studyno'])))
        # print(group[['wr_studyno','wr_nosmp']])
        study_group = group.groupby(['wr_studyno'], as_index=False)
        study_group = study_group['wr_nosmp']. agg(np.max)
        # print(sum(study_group['wr_nosmp']))
       # print(study_group)
        sample_list.append(sum(study_group['wr_nosmp']))
    y_sample.append(np.sum(sample_list))
```
